# Problem: Stock Simulation

- Create a stock Simulaiton
- Start with $100
- Have change random
- Record the returns, closing price, and amount of times price dipped below 0



```
import random
import numpy as np

def stock(ndays=252):

  dic = {'returns':[],'close':[],'below':0}
  start = 100

  while ndays > 0:
    
    ndays -= 1
    change = np.random.normal(loc=.5,scale=2,)
    close =  start * change
    returns = (close/start) - 1
    if close < 100:
      dic['below'] += 1    

    dic['returns'].append(returns) 
    dic['close'].append(close)

  return dic

```

```
stock()
```
